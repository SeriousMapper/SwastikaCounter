```python
import pandas as pd
import geocoder
df = pd.read_csv("scp_data_set.csv")
df_loc = pd.read_csv("geocoded_cities.csv")
print(df)
print(df_loc)
```

```python
lon_list = []
lat_list = []
found_address_list = []
i = 0
for ind in df.index:
    i += 1
    city = df['city'][ind]
    state = df['STATE'][ind]
    cityState = str(city) + "," + str(state)
    g = geocoder.bing(cityState, key='AnSJDRrLPJczP613_0oVzeYr4Uiai6BpbYMMjz3abpR1EJ7EAZ4qfv316QHW9JMv')
    if g.json != None:
        lon_list.append(g.json['lng'])
        lat_list.append(g.json['lat'])
        found_address_list.append(g.json['address'])
    else:
        lon_list.append(None)
        lat_list.append(None)
        found_address_list.append('Not Found')
    
    print(i ,' / ', len(df[['city']]))
```

```python

df['place'] = df['locations']
print(df.columns)
```

```python
new_df = df.copy()
new_df = new_df[['website', 'date of discovery or report','date', 'city', 'STATE',
       'source', 'reported phenomenon other than swastika',
       'reported phenomenon other than swastika 2',
       'reported phenomenon other than swastika 3', 'target', 'accompanying text',
       'accompanying visual signs', 'Nazi Reference', 'media', 'place',
       'category of place', 'structure', 'url_to_jpg']]
```

```python
print(lat_list)
```

```python
new_df.loc[:,"latitude"] = lat_list
new_df.loc[:,"longitude"] = lon_list
new_df.loc[:,"city_state"] = found_address_list
```

```python
new_df[['month', 'year']] = new_df['date'].str.split('-', expand=True)
```
